```python
import pandas as pd
import matplotlib.pyplot as plt

mydata=pd.read_csv("http://bit.ly/movieusers",sep="|",header=None)
print(mydata.head(20))
```

```python
mydata.columns=["col1","col2","col3","col4","col5"]
#print(mydata.col3)
mydata['col6']=mydata["col2"].map(str)+":"+mydata["col3"]
print(type(mydata))
print(mydata.columns.tolist())
```

```python
mydata.iloc[0:20].plot.bar(x='col4',y='col1',color='b',rot=90)
plt.show()
```

```python
print(mydata.groupby('col4').mean())
mydata.groupby('col4').mean().plot(y='col2', kind='bar', color='b')
print(mydata.col2.mean())
plt.show()
```
